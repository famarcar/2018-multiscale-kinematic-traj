```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Action classification

**[1]** The idea was to evaluate independently the features. In such plot, the result with other datasets is shown

**Figure to paper**

```python
# histogram per feature combination
results_1 = pd.read_csv("KTH definitive results - Exp_1.csv")
results_1.head(10)
```

```python
plt.figure(figsize=(12,5))

y_pos = np.array(range(len(results_1.Feature)))
plt.bar(y_pos, results_1.KTH,  alpha=0.7
        , width=0.3,  label="KTH")
plt.bar(y_pos+0.1, results_1.UT,alpha=0.6, width=0.3)
plt.bar(y_pos+0.2, results_1.Youtube,  alpha=0.7, width=0.3, align= 'edge')

plt.xticks(y_pos, results_1.Feature)

plt.legend(['KTH', 'UT',  "Youtube"])
plt.ylim([50, 100])
plt.tight_layout()

locs, labels = plt.xticks()

plt.setp(labels, rotation=45)


plt.xlabel("Kinematic features")
plt.ylabel("Accuracy ")
plt.title("Performance of kinematic features independently")
plt.grid(alpha=0.2)
plt.savefig("kinFeat.pdf")
plt.show()
```

####  histograms per feature

 Note. The features were computed with tree scales. The best feature was $\theta =$ **90.61**

**[2]** A combination of all features with theta was performed. The best combination was achieved with $\left \{ \theta, \kappa \right \} = $ **90.03**

**[3]** A combination of all features with $\left \{ \theta, \kappa \right \} $ was performed. The best combination was achieved with $\left \{ \theta, \kappa, a_T \right \} = $ **91.87**

**[4]** A combination of all features with $\left \{ \theta, \kappa,  a_T  \right\} $ was performed. The best combination was achieved with $\left \{ \theta, \kappa, a_T, N_y \right \} = $ **90.84**

```python
# histogram per feature combination
results_2 = pd.read_csv("KTHresultsAddfeat.csv")
results_2.head(10)
```

```python
plt.plot(results_2.features, results_2.feat,'o', c= "red", markeredgecolor="black",
         markersize= 7, alpha=0.6)
plt.plot(results_2.features, results_2["Theta+"],'o', c= "blue", markeredgecolor="black", 
         markersize= 7, alpha=0.6)
plt.plot(results_2.features, results_2["Theta+curv"],'o', c= "green", markeredgecolor="black",
         markersize= 7, alpha=0.6)
plt.plot(results_2.features, results_2["Theta + curv + acelT"],'o', c= "yellow", markeredgecolor="black",
         markersize= 7,  alpha=0.6)

plt.legend(['one F', 'Theta+',  "Theta+curv", "Theta+curv+acelT"])
plt.ylim([60, 95])
plt.tight_layout()


plt.legend(['KTH', 'UT',  "Youtube"])
plt.ylim([50, 100])
plt.tight_layout()

locs, labels = plt.xticks()

plt.setp(labels, rotation=45)


plt.xlabel("Kinematic features")
plt.ylabel("Accuracy ")
plt.title("Performance of add kinematic features")
plt.grid(alpha=0.2)
plt.savefig("AddkinFeat.pdf")
plt.show()
```

```python
# plot with different number of scales. 
results_temp = pd.read_csv("KTH-nyscales.csv")
results_temp.head(100)
```

```python
plt.plot(results_temp.time, results_temp.TCA_1S, 'o', color="blue", linewidth=2.0, alpha=0.3, linestyle="-", label="Accuracy")
plt.plot(results_temp.time, results_temp.TCA_3S, 'o', color="red", linewidth=2.0,  alpha=0.3,linestyle="-", label="Accuracy")

plt.plot(results_temp.time, results_temp.TCA_5S, 'o', color="orange", linewidth=2.0,  alpha=0.3,linestyle="-", label="Accuracy")

plt.plot(results_temp.time, results_temp.TCA_7S, 'o', color="green", linewidth=2.0,  alpha=0.3,linestyle="-", label="Accuracy")

plt.legend(['1S', '3S',  "5S", "7S"])
plt.grid(alpha=0.2)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.grid(alpha=0.2)
plt.ylim([30, 75])
plt.plot(results_temp.time, results_temp.TCA_1S, 'o', color="blue", linewidth=2.0, alpha=0.3, linestyle="-", label="Accuracy")
plt.plot(results_temp.time, results_temp.TCA_3S, 'o', color="red", linewidth=2.0,  alpha=0.3,linestyle="-", label="Accuracy")
plt.plot(results_temp.time, results_temp.TCA_5S, 'o', color="orange", linewidth=2.0,  alpha=0.3,linestyle="-", label="Accuracy")
plt.plot(results_temp.time, results_temp.TCA_7S, 'o', color="green", linewidth=2.0,  alpha=0.3,linestyle="-", label="Accuracy")
plt.legend(['1S', '3S',  "5S", "7S"])
plt.subplot(122)
plt.grid(alpha=0.2)
plt.ylim([75, 95])
plt.plot(results_temp.time, results_temp.TCA_1S, 'o', color="blue", linewidth=2.0, alpha=0.3, linestyle="-", label="Accuracy")
plt.plot(results_temp.time, results_temp.TCA_3S, 'o', color="red", linewidth=2.0,  alpha=0.3,linestyle="-", label="Accuracy")
plt.plot(results_temp.time, results_temp.TCA_5S, 'o', color="orange", linewidth=2.0,  alpha=0.3,linestyle="-", label="Accuracy")
plt.plot(results_temp.time, results_temp.TCA_7S, 'o', color="green", linewidth=2.0,  alpha=0.3,linestyle="-", label="Accuracy")
plt.legend(['1S', '3S',  "5S", "7S"])
```

```python
plt.plot(results_temp.time, results_temp.TCA_3S, 'o', 
         color="green", linewidth=2.0,  alpha=0.6,linestyle="-", markeredgecolor="black",
         markersize= 10)

plt.legend(['KTH three scales'])
plt.xlabel("% of video sequence")
plt.ylabel("Accuracy ")
plt.title("Temporal and partial recognition")
plt.grid(alpha=0.2)
plt.savefig("TempFeat.pdf")
plt.show()
```
